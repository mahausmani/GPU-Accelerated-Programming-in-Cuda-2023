2. Assume that we want to use each thread to calculate two (adjacent) elements 
of a vector addition. What would be the expression for mapping the thread/
block indices to i, the data index of the first element to be processed by a 
thread? 

    A. i=blockIdx.x*blockDim.x + threadIdx.x +2;

    B. i=blockIdx.x * threadIdx.x * 2;

    C. i=(blockIdx.x * blockDim.x + threadIdx.x)*2;

    D. i=blockIdx.x * blockDim.x*2 + threadIdx.x;


# Solution
We have launced ceil(1024/256) = 4 blocks and 128 threads per block, which gives us 512 total threads. This is half the dataset size. We are still able to iterate over all elements because each thread bow accesses two elements of the array. i = (global index)*2

So for global index = 0, i = 0

global index = 1, i = 2

global index = 2, i = 4


Hence we are launching half the number of threads and still are able to iterate the intira array

```python
%%cu 
#include <stdio.h>
#include <cuda.h>
#include <math.h>

__global__ void add_vector(int* a, int*c, int n){
    int i = (threadIdx.x + blockDim.x * blockIdx.x)*2;
	if (i<n)
		c[i] = a[i] + 1;
        c[i+1] = a[i+1] + 1;	
}
int main(){
    int *h_data = 0;
    int *d_data = 0;
    int *d_data_output = 0;
    int n = 1024;

    h_data = (int*)malloc(n*sizeof(int));

    cudaMalloc((void**)&d_data,n*sizeof(int));
    cudaMalloc((void**)&d_data_output,n*sizeof(int));

    for (int i = 0; i<n;i++){
        h_data[i] = i;
    }

    cudaMemcpy(d_data, h_data, n*sizeof(int), cudaMemcpyHostToDevice);

    add_vector<<<ceil(n/256),128>>>(d_data,d_data_output,n);

    for (int i = 0; i<n;i++){
            h_data[i] = 0;
        }

    cudaMemcpy(h_data, d_data_output, n*sizeof(int), cudaMemcpyDeviceToHost);

    for(int i = 0; i<n;i++){
        printf("h_data[%d] = %d\n", i,h_data[i]);
    }

    cudaFree(d_data);
    free(h_data);

    return 0;
}
```
